```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#libraries
```

```python
df=pd.read_csv('/content/table.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['ref_txn'].count() #example, counts all values except null
```

```python
difference_list=[]
i=1
while(i<=1000):
 df_1000=df.sample(frac=0.1) #644 entries approximately
 df_1=df_1000.sample(frac=0.5) #322 random entries in this dataframe
 df_2=df_1000.sample(frac=0.5) #322 random entries in this one
 tr_ratio_1=df_1['ref_txn'].count()/df_1['total_ref'].count() #calculating ratio
 tr_ratio_2=df_2['ref_txn'].count()/df_2['total_ref'].count() # """"
 print("ratio of df_1= "+str(tr_ratio_1))
 print("ratio of df_2= "+str(tr_ratio_2))
 tr_ratio_diff=tr_ratio_1-tr_ratio_2 #difference
 print("difference="+str(tr_ratio_diff))
 print("end of iteration number "+str(i))
 print()
 difference_list.append(tr_ratio_diff) #adding the ratio difference calculated in each iteration to a list
 i=i+1 #updating i
```

```python
diff_ser=pd.Series(difference_list)
```

```python
difference_df=diff_ser.to_frame(name='Difference')
```

```python
difference_df.shape
```

```python
difference_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.hist(difference_df.Difference)
plt.title('Frequency Distribution')
plt.xlabel('Differences obtained')
```
